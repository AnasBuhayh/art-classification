### Reading images

```python
import random
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg # for loading in images

%matplotlib inline
```

```python
# reading styles
import os

# get the styles names
styles = os.listdir('archive')

styles_lables = list()

# add target label to the styles
for index, style in enumerate(styles):
    styles_lables.append((style, index))

print(styles)
print(styles_lables)
```

```python
import cv2 

# This function should take in an RGB image and return a new, standardized version
def standardize_input(image):
    image_crop = np.copy(image)
    row_crop = 7
    col_crop = 8
    image_crop = image[row_crop:-row_crop, col_crop:-col_crop, :]
    ## TODO: Resize image and pre-process so that all "standard" images are the same size
    standard_im = cv2.resize(image_crop, (128, 128))
    return standard_im
```

```python
# test image
test_img = glob.glob(f'archive/Western_Medieval/Western_Medieval/9223372032559869461.jpg')[0]
test_img = mpimg.imread(test_img)
print(test_img.shape)
plt.imshow(test_img)
```

### Preparing the data

```python
#standarize the image
std_img = standardize_input(test_img)
print(std_img.shape)
plt.imshow(std_img)
```

```python
# reading samples and adding lables
ims_list = list()

for style in styles_lables:

    for file in glob.glob(f'archive/{style[0]}/{style[0]}/*')[:10]:
        # Read in the image
        im = mpimg.imread(file)

        if len(im.shape) < 3: #skip black and white images
            continue
        
        # standarize(im) 28x28
        std_im = standardize_input(im)
        
        # normalize(im) im/255
        norm_std_im = std_im/255
                
        # Append the image to the green image list
        ims_list.append((norm_std_im,style[1]))
```

```python
#check the images count
len(ims_list)
```

```python
# randomize the images
import random
random.seed(31101992)

random.shuffle(ims_list)
```

```python
# split train and test labels
X = [img[0] for img in ims_list]
y = [label[1] for label in ims_list]
```

```python
print(len(X))
print(len(y))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=31, stratify=y
)
```

### Setting up the model

```python
import tensorflow as tf
```

```python
# Check if AMD GPU is available
print("AMD GPU is", "available" if tf.config.list_physical_devices('GPU') else "not available")
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(128, 128, 3)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
training_history = model.fit(X_train, y_train, epochs=10)
```
